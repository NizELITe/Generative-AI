# to work with hugging face (free of cost)

```python
import os
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import persist
from langchain.document_loaders import PyPDFLoader

# Load the PDF document
def extract_text_from_pdf(pdf_path):
    loader = PyPDFLoader(pdf_path)
    pages = loader.load_and_split()
    text_chunks = [page.text for page in pages]
    return text_chunks

# Path to your PDF document
pdf_path = '/content/LAND_RESTORATION_raw_Laung.pdf'
document_texts = extract_text_from_pdf(pdf_path)

# Initialize Sentence Transformer model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Create embeddings for each text chunk
embeddings = model.encode(document_texts)

# Initialize Chroma vector database
persist_directory = '/content'
client = chromadb.Client()

# Create or load collection
vectordb = client.create_collection(name='document_collection') if not os.path.exists(persist_directory) else client.load_collection(name='document_collection')

# Add documents to the collection
vectordb.add(
    documents=document_texts,
    embeddings=embeddings
)

# Persist the database
persist(client, persist_directory)

# Check the number of documents
print(f"Number of documents in the collection: {vectordb.count()}")
```

```python
# Perform similarity search
question = "Your search query here"
question_embedding = model.encode([question])[0]

# Search the vector database for similar documents
results = vectordb.similarity_search(query_embedding=question_embedding, k=5)

# Print results
for result in results:
    print(f"Document Text: {result['document']}, Similarity: {result['similarity']}")
    print("----------")

print("Similarity search completed.")
```
